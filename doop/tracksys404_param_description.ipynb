```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
root= ET.parse('data/tracksysParam.xml')
```

```python
for elem in root.findall('.//Parameter/Description'):
    pass
    #print(elem.attrib.get('parameterName')) #'.//Parameter'
    #print(elem.text) #'.//Parameter/Description'
```

```python
list_paramName= [elem.attrib.get('parameterName') for elem in root.findall('.//Parameter')]
```

```python
list_description= [elem.text for elem in root.findall('.//Parameter/Description')]
```

list to df

```python
df= pd.DataFrame({'parameterName':list_paramName, 'description':list_description})
```
