```python
x= symbols('x')
y= symbols('y')
z= symbols('z')
```

```python
expr= x+ 2*y
print(expr)
print("__class__: ",expr.__class__)
print("args: ", expr.args)
print("subs{1,2}: ",expr.subs({x:1,y:2}))
print("subs{1,2}.evalf: ", expr.subs({x:1, y:2}).evalf())
```

#### 예쁘게 출력 pprint
pretty print

```python
from sympy import sympify, pprint, init_printing, symbols
```

```python
y= symbols('y')
init_printing(order= 'rev-lex')
print(y**4+ y**3)
pprint(y**4+ y**3)
```

```python
init_printing(order= 'grevlex')
pprint(y**2+ y**3)
```
